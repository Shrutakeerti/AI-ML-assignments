```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv(input("/content/diabetes(in).csv"))

# Function to perform bootstrap sampling
def bootstrap_split(data):
    train_indices = np.random.choice(data.index, size=len(data), replace=True)
    return data.loc[train_indices], data.loc[~data.index.isin(train_indices)]

# Function to calculate bootstrap means
def bootstrap_means(data, column, n_samples):
    return np.array([bootstrap_split(data)[0][column].mean() for _ in range(n_samples)])

# Perform bootstrap and calculate means
train, test = bootstrap_split(data)
means = bootstrap_means(train, 'Glucose', 1000)

# Display results
print(f"Training set size: {len(train)}, Test set size: {len(test)}")
print(f"Mean of bootstrap means: {means.mean()}, Std dev: {means.std()}")

# Plot histogram
plt.hist(means, bins=30, edgecolor='k', alpha=0.7)
plt.title('Bootstrap Means Distribution (Glucose)')
plt.show()
```
